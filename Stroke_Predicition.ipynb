#Menghapus outlier

```python
import numpy as np #Mengimpor library NumPy
q11,q31 = np.percentile(df1['bmi'],[25,75]) #menghitung kuartil ke-1 dan kuartil ke-3 dari kolom 'bmi' dalam DataFrame 'df1'
s1 = q31-q11 #Menghitung rentang antara kuartil ke-1 dan kuartil ke-3
ba1= q31+(1.5*s1)#Menghitung batas atas untuk  menentukan batasan outlier
bw1=q11-(1.5*s1)#Menghitung batas bawah untuk  menentukan batasan outlier
print(q11) #Menampilkan nilai dari kuartil ke-1
print(q31) #Menampilkan nilai dari kuartil ke-3
print(s1) #Menampilkan nilai dari interquartile range
print(ba1) #Menampilkan nilai dari batas atas
print(bw1) #Menampilkan nilai dari batas bawah
```

```python
data1 = df1[(df1['bmi'] < bw1) | (df1['bmi'] > ba1)]# Data akan disaring hanya jika nilai dalam kolom 'bmi' kurang dari 'bw1' (batas bawah) atau lebih besar dari 'ba1' (batas atas)
data1
```

```python
from scipy.stats import zscore

# Menghitung z-score dari kolom 'bmi'
z_scores = zscore(df1['bmi'])

# Menentukan batas z-score untuk outlier
threshold = 2

# Mengidentifikasi dan menghapus outlier
data_no_outliers = df1[(np.abs(z_scores) < threshold)]
```

```python
plt.figure(figsize=(8, 6)) #Membuat figur dengan ukuran 8x6 inci menggunakan matplotlib
sns.boxplot(y=data_no_outliers['bmi']) #Menggambar boxplot dari kolom 'bmi'
plt.title('Boxplot BMI Tanpa Outlier') #Memberikan judul pada boxplot
plt.xlabel('BMI') #Memberikan label pada sumbu x
plt.show() #Menampilkan boxplot
```
